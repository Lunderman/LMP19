```python
import sys
sys.path.insert(0, '../')
import LMP19
import numpy as np
import matplotlib.pyplot as plt
```

# Lorenz `63 example
* Observe only $x_0$ and $x_2$.

```python
L63 = LMP19.Models.L63()          

H = np.array([[1,0,0],[0,0,1]])   # x[1] not observed
nSteps = 500
Y = L63.get_data(nSteps=nSteps)   #truth
y = H@Y                           #observed truth

EnKF = LMP19.Estimation.Sqrt_EnKF(L63.M,nDim=3,H=H,inflation=0.25)

nE = 10
X_a = np.zeros((3,nE,nSteps))
Spread = np.zeros(nSteps)
RMSE = np.zeros(nSteps)

for ee in range(nE):
    X_a[:,ee,0] = np.ravel(L63.get_data(nSteps=1))

for kk in range(1,nSteps):
    X_a[:,:,kk] = EnKF.run(X_a[:,:,kk-1],y[:,kk])
    Spread[kk] = (np.trace(np.cov(X_a[:,:,kk]))/3)**0.5
    RMSE[kk] = np.sqrt(np.mean((np.mean(X_a[:,:,kk],axis=1)-Y[:,kk])**2))
```

```python
for kk in range(3):
    for ee in range(nE):
        plt.plot(X_a[kk,ee,:],'C0')
    plt.plot(Y[kk,:],'C1',linewidth=2)
    plt.show()
```

```python
plt.plot(RMSE,label='RMSE')
plt.plot(Spread,label='Spread')
plt.xlabel('Time')
plt.legend()
plt.show()
```
